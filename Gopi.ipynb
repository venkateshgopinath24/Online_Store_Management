```python
class Shopping:
    
    # init method
    def __init__(self):
        self.products = pd.read_csv("products.csv")
        self.cart_items = {}
        self.df = pd.DataFrame()
        self.products_dummy = pd.DataFrame()
        self.call()
        
    def call(self):
        
        a = input("Select among following options: \nList\nCart\nAdd\nRemove\nCheckout\nExit\n")
        if(a == 'List'):
            self.inventory_list()
        elif(a == "Cart"):
            self.cart()
        elif(a == "Add"):
            self.add_to_cart()
        elif(a == "Remove"):
            self.remove_from_cart()
        elif(a == "Checkout"):
            self.checkout()
        elif(a == "Exit"):
            self.exit()
        
    def inventory_list(self):
        
        self.products_dummy = copy(self.products)
        
        display(self.products_dummy)
        
        return self.call()
        
        
        
    def cart(self):
        
        display(self.df)
      
        
        return  self.call()
        
    def add_to_cart(self):
        
        
        #collection input from user
        p1, q1 = input("Enter the product name and qty with , between them: ").split(",")
        print(p1)
        print(q1)
        #write condition to check inventory
        if (self.products_dummy[self.products_dummy["Name"] == p1]["Qty"] >= int(q1)).all():
            
            self.cart_items[p1] = q1
            
        else:
            print("Quantity is more than what is available in inventory")
        
        
        #empty list to note prices of items in cart
        price = []
        
        #calculationg prices for items in cart(Multiplying unit price and quantity)
        for key, value in self.cart_items.items():
            p = int(value) * int(self.products_dummy[self.products_dummy["Name"] == key]["Price"])
            price.append(p)
        
        
        #creating dataframe to display cart
        self.df = pd.DataFrame(list(self.cart_items.items()))
        
        #adding column names
        self.df.columns = ['Name', 'Qty']
        
        #Creating prices column
        self.df['Total_Price'] = price
            
        
        
        return self.call()
             
        
    def remove_from_cart(self):
        Print("Which item do you want to remove?")
        display(self.cart_items)
        
        #take input from user
        i = input("Enter item name: ")
        
        #try catch for item in cart
        try:
            del self.cart_items[i]
        except KeyError:
            print("Item not in cart")
            
        self.call()
        
    def checkout(self):
        #Calculating subtotal
        tp = self.df['Total_Price'].sum()
        print("subtotal is: ", tp)
        
        #calculating tax amount
        tax = tp * 0.7
        print("Tax amount is: ", tax)
        
        #calculating total
        tp *= 1.07
        print("Total price(Including tax) is: ", tp)
        
        #remove item from dataframe        
        for i in self.df["Name"]:
            
            #getting index for each item in 'df' from 'products_dummy' data frame 
            idx1 = self.products_dummy.index[self.products_dummy["Name"] == i]
            
            #getting index for each item in 'df'
            idx2 = self.df.index[self.df["Name"] == i ]
            
            #Subtracting qty of "df" from qty of "Products_dummy" 
            self.products_dummy.iloc[idx1,2] -= int(self.df.iloc[idx2]["Qty"])
        
        self.call()
        
        
    def exit(self):
        
        for i in self.df["Name"]:
            
            #getting index for each item 'products_dummy' data frame 
            idx1 = self.df.index[self.df["Name"] == i ]
            print(idx1)

            #getting index for each item in 'products'
            idx2 = self.products.index[self.products["Name"] == i ]
            
            print(idx2)

            #Subtracting qty of "df" from qty of "Products_dummy" 
            self.products.iloc[idx2,2] = self.products.iloc[idx2,2] - int(self.df.iloc[idx1]["Qty"])
        self.products.to_csv("products.csv", index = False)
            
    
```
